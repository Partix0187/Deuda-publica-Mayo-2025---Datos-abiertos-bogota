```python
import pandas as pd
import re
```

```python
file_path = "/content/deuda-mes-mayo-2025-deuda-publica.csv"
```

```python
deuda_publica = pd.read_csv(file_path, sep=';')
```

```python
print("Datos cargados inicialmente:")
deuda_publica.head()
```

```python
deuda_publica.info()
```

#Limpieza y estandarización

##Estandarizar nombres de columnas

```python
# Crear una copia del DataFrame para no modificar el original
deuda_pu_limpio = deuda_publica.copy()

# Limpiar los nombres de las columnas
columnas_nuevas = {
    'Código Ministerio Hacienda': 'codigo_mh',
    'Entidad Acreedora': 'entidad_acreedora',
    'No. Contrato': 'no_contrato',
    'Saldo adeudado Capital 30-Abr-2025': 'saldo_cap_abr_2025',
    'Ajustes': 'ajustes',
    'Amortización': 'amortizacion',
    'Intereses': 'intereses',
    'Comisiones': 'comisiones',
    'Saldo Adeudado Capital 31-May-2025': 'saldo_cap_may_2025'
}

deuda_pu_limpio = deuda_pu_limpio.rename(columns=columnas_nuevas)

print("DataFrame con columnas renombradas:")
deuda_pu_limpio.head()
```

##Convertir Columnas Numéricas

```python
# Lista de columnas que debemos convertir a tipo numérico
columnas_numericas = [
    'saldo_cap_abr_2025',
    'ajustes',
    'amortizacion',
    'intereses',
    'comisiones',
    'saldo_cap_may_2025'
]

# Iterar sobre cada columna para limpiarla y convertirla
for col in columnas_numericas:
    deuda_pu_limpio[col] = deuda_pu_limpio[col].astype(str).str.replace('.', '', regex=False).astype(float)

print("\nTipos de datos después de la conversión:")
deuda_pu_limpio.info()

print("\nVista previa de los datos con valores numéricos:")
deuda_pu_limpio.head()
```

##Estandarizar los Nombres de las Entidades Acreedoras

```python
def limpiar_nombre_entidad(nombre):
    if not isinstance(nombre, str):
        return 'DESCONOCIDO'

    # 1. Normalizar: quitar tildes y convertir a mayúsculas
    nombre_limpio = nombre.upper()
    tildes = {'Á': 'A', 'É': 'E', 'Í': 'I', 'Ó': 'O', 'Ú': 'U'}
    for acento, sin_acento in tildes.items():
        nombre_limpio = nombre_limpio.replace(acento, sin_acento)

    nombre_limpio = re.split(r'\s+-\s+|\s+\d+|\s+(?:I|II|III|IV|V)$', nombre_limpio)[0]

    # 3. Quitar espacios extra al inicio o al final
    return nombre_limpio.strip()

deuda_pu_limpio['entidad_limpia'] = deuda_pu_limpio['entidad_acreedora'].apply(limpiar_nombre_entidad)


# Mostrar la columna original y la nueva columna limpia para comparar
# Usaremos .unique() para ver todos los nombres únicos generados y confirmar la limpieza
print("Nombres de entidades únicos después de la limpieza mejorada:")
print(sorted(deuda_pu_limpio['entidad_limpia'].unique()))

print("\nEjemplo de comparación:")
# Mostramos algunos ejemplos para ver la transformación
print(deuda_pu_limpio[['entidad_acreedora', 'entidad_limpia']].tail(15))
```

##Calcular el Saldo Total Adeudado General

```python
# Agrupar por la entidad limpia y sumar el saldo de mayo 2025
deuda_por_entidad = deuda_pu_limpio.groupby('entidad_limpia')['saldo_cap_may_2025'].sum().reset_index()

# Ordenar los resultados de mayor a menor deuda
deuda_por_entidad = deuda_por_entidad.sort_values(by='saldo_cap_may_2025', ascending=False)

# Para mejorar la legibilidad, formateamos el número
pd.options.display.float_format = '{:,.2f}'.format

print("\n--- SALDO ADEUDADO TOTAL POR ENTIDAD (MAYO 2025) ---")
print(deuda_por_entidad)
```

```python
# Filtrar el DataFrame para eliminar las filas donde 'entidad_limpia' es 'DESCONOCIDO'
deuda_pu_limpio_filtrado = deuda_pu_limpio[deuda_pu_limpio['entidad_limpia'] != 'DESCONOCIDO'].copy()

# Verificamos que se hayan eliminado las filas
print("DataFrame después de eliminar 'DESCONOCIDO':")
print(deuda_pu_limpio_filtrado['entidad_limpia'].unique())

# Actualizar la variable deuda_por_entidad con los datos filtrados
deuda_por_entidad_filtrado = deuda_pu_limpio_filtrado.groupby('entidad_limpia')['saldo_cap_may_2025'].sum().reset_index()
deuda_por_entidad_filtrado = deuda_por_entidad_filtrado.sort_values(by='saldo_cap_may_2025', ascending=False)

print("\n--- SALDO ADEUDADO TOTAL POR ENTIDAD (MAYO 2025) - SIN DESCONOCIDO ---")
display(deuda_por_entidad_filtrado)
```

```python
# Define the output file path
output_file_path = "/content/deuda_total_por_entidad_limpio.csv"

# Save the filtered DataFrame to a CSV file
deuda_por_entidad_filtrado.to_csv(output_file_path, index=False)

print(f"DataFrame guardado exitosamente en: {output_file_path}")
```
